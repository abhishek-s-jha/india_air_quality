```python
import json
import pandas as pd
from pprint import pprint
%config Completer.use_jedi = False
```

```python
# with open('stations_geo_loc_data','w') as fp:
#     fp.write(data)
    
with open('stations_geo_loc_data', 'r') as fp:
    data = fp.read()
data = json.loads(data, strict=False)
```

```python
all_stations = data['map']['station_list']
print("Total Station :: ", len(all_stations))
```

```python
for station in all_stations:
    print('Station ID :: ', station['station_id'] )
    print('\tStation Name          :: ', station['station_name'])
    print('\tStation Type          :: ', station['station_type'])
    print('\tVendor                :: ', station['vendor'])
    print('\t(Latitude, Longitude) :: ', (float(station['latitude']), float(station['longitude'])))
    break
```

```python
processed_station_geo_loc = {}
csv_dump = []
for idx,station in enumerate(all_stations):
    if station['station_id'] not in processed_station_geo_loc.keys():
        processed_station_geo_loc[station['station_id']] = {'name':station['station_name'],
                                                            'latitude':float(station['latitude']),
                                                            'longitude':float(station['longitude'])
                                                           }
        csv_dump.append([station['station_id'], 
                        station['station_name'],
                        float(station['latitude']), 
                        float(station['longitude'])]
                       )
        
```

```python
with open('processed_station_geo_loc.json','w') as fp:
    json.dump(processed_station_geo_loc, fp)
```

```python
pprint(processed_station_geo_loc)
```

```python
# id, name, latitude, longitude
df = pd.DataFrame(data=csv_dump, columns=['id', 'name', 'latitude', 'longitude'])
df.to_csv('processed_station_geo_loc.csv', header=True, index=False)
```

```python
import glob
```

```python
all_files = glob.glob("./processed_stations_establishment_timeline/*/*")
```

```python
region_info = {}
csv_dump = []
for file in all_files:
    df_1 = pd.read_csv(file)
    for row in df_1.values:
        state = row[1]
        city = row[2]
        station_name = row[3]
        if station_name not in region_info.keys():
            region_info[station_name] = [state, city]
            csv_dump.append([station_name, state, city])
#     break
```
