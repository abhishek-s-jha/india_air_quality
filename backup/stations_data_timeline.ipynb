```python
from glob import glob
import os
import pandas as pd
from datetime import datetime
%config Completer.use_jedi=False
```

```python
df_1 = pd.read_csv('./processed_station_geo_loc.csv')
df_2 = pd.read_csv('./stations_region.csv')
df = pd.merge(df_1, df_2, on='name')
df.head()
```

```python
df.sort_values(['state', 'city'], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
root_dir = "./stations_start_info"
start_date_data = []
for idx in range(len(df)):
    
    state = df.state[idx]
    city = df.city[idx]
    station_id = df.id[idx]
    station_name = df.name[idx]

    print(f"{idx}. {state} :: {city} :: {station_id} :: {station_name} :: ", end='')

    file_path = os.path.join(root_dir, state, city + '_' + station_id + '.csv')

    start_date_df = pd.read_csv(file_path).T
    start_date_df = start_date_df.set_axis(start_date_df.iloc[0].astype(int), axis=1).drop(start_date_df.index[0])
    start_date_df.dropna(axis=1, how='all', inplace=True)
    
    year_col = start_date_df.columns[-1]
    start_month = start_date_df[year_col].first_valid_index()
    month_number = datetime.strptime(start_month, '%B').month
    
    start_date_data.append([start_month,month_number, year_col])
#     break
```

```python
start_date_data
```

```python
df_1 = pd.DataFrame(start_date_data, columns=['start_month', 'start_month_num', 'year'])
```

```python
final_df = pd.concat([df, df_1], axis=1)
final_df
```
