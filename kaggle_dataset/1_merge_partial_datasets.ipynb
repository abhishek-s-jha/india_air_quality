```python
import os
import pandas as pd

%config Completer.use_jedi=False
```

```python
loc = "./stations_data/Haryana/site_5051/"
os.listdir(loc)
```

```python
df = pd.read_csv(os.path.join(loc, 'Sirsa_site_5051_0.csv'))
for file in sorted(os.listdir(loc))[1:]:
    print(file)
    new_df = pd.read_csv(os.path.join(loc, file))
    
    df = pd.concat([df,new_df])
```

```python
df
```

```python
file_loc = os.path.join('/'.join(loc.split('/')[:-2]), loc.split('/')[-3] + '_'+ loc.split('/')[-2] + '.csv')
df.to_csv(file_loc,header=True, index=False)
```
