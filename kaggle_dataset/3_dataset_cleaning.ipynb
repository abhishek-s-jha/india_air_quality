```python
import os
import pandas as pd
from tqdm import tqdm_notebook
from glob import glob
from datetime import datetime

%config Completer.use_jedi=True
```

```python
cleared_data_loc = "./cleaned_stations_data/*/*"
```

```python
df_columns = ['id', 'state', 'city', 'start_month', 'start_month_num' ,'start_year', 'file_name']
df_data = []

for file in tqdm_notebook(glob(cleared_data_loc)):
#     print(file)
    df = pd.read_csv(file)
    
    site_id = '_'.join(file.split('/')[-1][:-4].split('_')[1:])
    state = file.split('/')[2].strip()
    city = file.split('/')[-1].split('_')[0].strip()


    # Assuming you have the string stored in a variable named 'date_string'
    date_string = df.values[0][0]
    # Convert the string to a datetime object
    datetime_obj = datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')
    # Extract the month name from the datetime object
    start_month = datetime_obj.strftime('%B')
    start_month_num = datetime_obj.month
    start_year = datetime_obj.year

    file_name = state_abbreviations[state]
    
    df_data.append([site_id, state, city, start_month, start_month_num ,start_year, file_name])


#     break
```

```python
new_stations_details = pd.DataFrame(df_data, columns=df_columns)
all_states = list(set(new_stations_details.state.values))
```

```python
new_col = ['id', 'file_name']
new_data = []
for st in all_states:
#     st='Karnataka'
    x = new_stations_details[new_stations_details['state'] == st].sort_values(by=['start_year', 'start_month_num'])
    
    # Generate the sequential number
    x['sequential_number'] = x.groupby('file_name').cumcount() + 1

    # Convert the sequential number to a 3-digit string with leading zeros
    x['sequential_number'] = x['sequential_number'].apply(lambda x: str(x).zfill(3))

    # Generate the new column by combining 'file_name' and 'sequential_number', prefixed with 'A'
    x['new_column'] = state_abbreviations[st] + x['sequential_number']
    
    for row in x.values:
        new_data.append([row[0], row[-1]])
#     break
```

```python
new_df = pd.DataFrame(new_data, columns= new_col)
```
